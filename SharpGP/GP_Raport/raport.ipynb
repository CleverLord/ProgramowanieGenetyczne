```python
import json
import matplotlib.pyplot as plt
from statistics import mean as avg
data  = json.load(open('TestSet_Final_3_5000.json'))
print(data.keys())
avg_fit = [avg([number for number in generacja['fitnesses'] if number != 'Infinity']) for generacja in data['generations']]
plt.plot(avg_fit[0:4])
plt.show()
plt.ylim([0, 20000])
plt.plot(avg_fit[4:])
```

```python
avg_depth = [avg([number for number in generacja['populationDepths']]) for generacja in data['generations']]
min_depth = [min([number for number in generacja['populationDepths']]) for generacja in data['generations']]
max_depth = [max([number for number in generacja['populationDepths']]) for generacja in data['generations']]
plt.plot(avg_depth,color='orange')
plt.plot(min_depth,color='cyan')
plt.plot(max_depth,color='red')
plt.show()
```

```python
number_of_best_programs = [generacja['fitnesToUnits'][list(generacja['fitnesToUnits'].keys())[0]] for generacja in data['generations']]
# for x in data['generations']:
#     print(type(x['fitnesToUnits'].keys()))
plt.plot(number_of_best_programs)
```
